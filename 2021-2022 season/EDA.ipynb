```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('Arsenal season 2021_2022 - Sheet1.csv')
```

```python
df.columns
```

```python
drop=['Prediction','GF_prediction', 'GA_prediction','Predicted_formation','pred_gk_Ratings','pred_def_ratings', 'pred_mid_ratings', 'pred_forw_ratings']
```

```python
df=df.drop(drop,axis=1)
```

```python
df=df.iloc[:16]
```

```python
df['Formation'].replace('3-4-2003','3-4-3',inplace=True)
```

```python
df
```

```python
result=pd.get_dummies(df['Result'])
```

```python
venue=pd.get_dummies(df['Venue'])
```

```python
df=pd.concat([df,result,venue],axis=1)
```

```python
df.drop('Result',axis=1,inplace=True)
```

```python
df.drop('Venue',axis=1,inplace=True)
```

# Let's see what correlates highly to result of the match

```python
plt.figure(figsize=(12,6))
df.corr()['W'].sort_values()[:-1].plot(kind='barh',color='green')

plt.title('Numerical categories correlation with match results')
```

```python
plt.figure(figsize=(12,6))
df.corr()['L'].sort_values()[:-1].plot(kind='barh',color='red')

plt.title('Numerical categories correlation with match results')
```

```python
plt.figure(figsize=(12,6))
df.corr()['D'].sort_values()[:-1].plot(kind='barh',color='yellow')

plt.title('Numerical categories correlation with match results')
```

    Findings:
        1. Arsenal midfields contribute most to winning matches
        
        2. Expected goals (XG) does not result in winning matches. So, we cannot solely say that team with a higher XG has a hgiher chance in winning games.
        
        3. Playing home is also a very important factors that contribute to Arsenal winning matches. 

# How does the player perform over time?

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 9))

axes[0,0].plot(df['act_gk_ratings'],'tab:orange')
axes[0,0].set_title("Goalkeeper ratings after 16 matches")

axes[0,1].plot(df['act_def_ratings'],'tab:red')
axes[0,1].axis('tight')
axes[0,1].set_title("Defenders ratings after 16 matches")

axes[1,0].plot(df['act_mid_ratings'])
axes[1,0].axis('tight')
axes[1,0].set_title("Midfielders ratings after 16 matches")


axes[1,1].plot(df['act_forw_ratings'],'tab:green')
axes[1,1].axis('tight')
axes[1,1].set_title("Forwards ratings after 16 matches");


for axes in axes.flat:
    axes.set(xlabel='Match', ylabel='Ratings')

fig.tight_layout()
```

    We can conclude here that for the past matches, only the goalkeeper (Ramsdale) has been consistent in performing.
